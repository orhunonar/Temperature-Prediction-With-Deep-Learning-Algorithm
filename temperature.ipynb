```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
celcius_degree = np.array([-40, -10, 0, 8, 15, 22, 38],dtype=float)
fahrenheit_degree = np.array([-40, 14, 32, 46, 59, 72, 100],dtype=float)
```

```python
def TrainValidateModel():
    model = keras.Sequential() # Sequential is a linear stack of layers
    model.add(Dense(1,input_shape=[1])) # input_shape is the shape of the input data
    model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(0.1)) # Adam is a good optimizer
    model.fit(celcius_degree, fahrenheit_degree, epochs=500,verbose=False) # verbose=False is to avoid printing the progress of the training
    
    ValidateModel(model,celcius_degree,fahrenheit_degree) # Validate the model

    return model
```

```python
def ValidateModel(model,celcius_degree,fahrenheit_degree):
    predictions = model.predict(celcius_degree)
    for i in range(len(predictions)):
        print("Celcius: %d, Fahrenheit: %d" % (celcius_degree[i], fahrenheit_degree[i])) # Print the real and predicted values
        print("Predicted: %d" % (predictions[i])) # Print the predicted value
        print("Error: %d" % (abs(predictions[i] - fahrenheit_degree[i]))) # Print the error
        print("")
```

```python
def PredictCelcius(model,celcius_degree):
    return model.predict(np.array([celcius_degree]))   # Predict the fahrenheit degree for the given celcius degree
```

```python
model=TrainValidateModel() # Train the model
```

```python
prediction_celcius=int(input("Enter the temperature in Celcius: ")) # Get the celcius degree from the user
result=PredictCelcius(model,prediction_celcius) # Predict the fahrenheit degree for the given celcius degree
print(result)
```
